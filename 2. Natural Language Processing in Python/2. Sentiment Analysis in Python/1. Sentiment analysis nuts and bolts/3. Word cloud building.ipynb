# Word cloud building

## What is sentiment analysis?

* Sentiment analysis is the process of understanding the opinion of an author about a subject.

## Why word clouds?

* Pros:

    * can reveal the essential
    
    * provide an overall sense of the text
    
    * easy to grasp and engaging

* Cons:

    * sometimes confusing and uninformative
    
    * with larger text, require more work

## Code of building a word cloud in Python:

````python
from wordcloud import WordCloud

two_cities = 'It was the best of times, it was the worst of times, \
it was the age of wisdom, it was the age of foolishness, \
it was the epoch of belief, it was the epoch of incredulity, \
it was the season of Light, it was the season of Darkness, \
it was the spring of hope, it was the winter of despair, \
we had everything before us, we had nothing before us, \
we were all going direct to Heaven, we were all going direct the other way – \
in short, the period was so far like the present period, \
that some of its noisiest authorities insisted on its being received, \
for good or for evil, in the superlative degree of comparison only.'

cloud_two_cities = WordCloud().generate(two_cities)

# How does cloud_two_cities look like?
cloud_two_cities
````

## What are the essential parameters for personalizing a word cloud function?

* Background color(`background_color`).

* Size(`width` and `height`) and font of the words(`min_font_size` and `max_font_size`), scaling(`scale` and `relative_scaling`).

* Stopwords(`stopwords`).

## Code of displaying the word cloud:

````python
import matplotlib.pyplot as plt

plt.imshow(cloud_two_cities, interpolation='bilinear')
plt.axis('off')
plt.show()
````

## Practice question for querying which words that are in the word cloud:

* A string called `illuminated` contains an emotionally charged quote from the book '*Everything is Illuminated*' by Jonathan Safran Foer. Generate and display a word cloud using the `illuminated` string.

* Which of the following answers contains only words that could be seen in the word cloud?

    $\Box$ Sad, other, happy.
    
    $\Box$ Repeat, conversation , convince.
    
    $\boxtimes$ Day, worse, others.
    
    $\Box$ Know, thing, everything.

$\blacktriangleright$ **Package pre-loading:**

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

$\blacktriangleright$ **Data pre-loading:**

````python
illuminated = 'I am not sad, he would repeat to himself over and over, \
I am not sad. As if he might one day convince himself or convince others -- \
the only thing worse than being sad is for others to know that you are sad. \
I am not sad.'
````

$\blacktriangleright$ **Question-solving method:**

````python
cloud_illuminated = WordCloud().generate(illuminated)
plt.imshow(cloud_illuminated, interpolation='bilinear')
plt.axis('off')
plt.show()
````

## Practice exercises for word cloud building II:

$\blacktriangleright$ **Package pre-loading:**

````python
import pandas as pd
import matplotlib.pyplot as plt
````

$\blacktriangleright$ **Data pre-loading:**

````python
descriptions = open('ref4. A catenation of 100 movie reviews.txt').read()
my_stopwords = set(
    pd.read_csv('ref5. Movie-specific stopwords.csv',
                header=None)[0].to_list())
````

$\blacktriangleright$ **Movie reviews' word cloud practice:**

````python
# Import the word cloud function
from wordcloud import WordCloud

# Create and generate a word cloud image
my_cloud = WordCloud(background_color='white',
                     stopwords=my_stopwords).generate(descriptions)

# Display the generated wordcloud image
plt.imshow(my_cloud, interpolation='bilinear')
plt.axis("off")

# Don't forget to show the final image
plt.show()
````
